```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Thinkful-Ed/data-science-lectures/master/holiday_shift_dataset.csv")
```

```python
df
```

```python
df_deep=df.copy()
```

```python
df_deep
```

```python
df_deep['num_employees'].replace(40, 26, inplace=True)
```

```python
df_deep
```

```python
df
```

```python
df_shallow=df.copy(deep=False)
```

```python
df_shallow
```

```python
df_shallow['num_employees'].replace(40, 26, inplace=True)
```
