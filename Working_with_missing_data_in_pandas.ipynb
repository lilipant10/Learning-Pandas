```python
import pandas as pd
df_prod=pd.read_csv('https://raw.githubusercontent.com/Thinkful-Ed/data-science-lectures/master/product_information.csv')
df_prod.head()
```

```python
df_prod.dropna()
```

```python
df_prod.fillna(50)
```

```python
prod_qty_median=df_prod['prod_qty'].median()
df_prod.fillna(prod_qty_median)
```

```python
df_prod.fillna(df_prod['prod_qty'].median())
```

```python
#mode
prod_price_mode=df_prod['prod_price'].mode()
df_prod.replace(0, prod_price_mode)
```

```python
df_prod.replace(0, df_prod['prod_price'].mode()[0])
```

```python
df_prod.interpolate(method='linear')
```

```python
#methopd='pad'
df_prod.interpolate(method='pad')
```
