```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Thinkful-Ed/data-science-lectures/master/loss_prevention_data.csv")
df.head()
```

```python
df['item_cost']==399.99
```

```python
df['item_cost']<=10.00
```

```python
df_under_10=df['item_cost']<=10.00
df[df_under_10]
```

```python
df_2_june= df['date']=='2-june'
df_electronics=df['item_dept']=='electronics'
df[df_2_june]& df_electronics
```

```python
df_electronics= df['item_dept']=='electronics'
df_over_100 = df['item_cost'] >= 100.00
df[df_electronics|df_over_100]
```
