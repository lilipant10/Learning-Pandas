```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("https://raw.githubusercontent.com/Thinkful-Ed/data-science-lectures/master/emp_avg_data.csv")
df
```

```python
sns.lineplot(x='avg_emp_rating',y='avg_emp_tenure_in_years',data=df)
plt.show()
```

```python
sns.lineplot(x='avg_emp_rating',y='avg_emp_tenure_in_years', data=df)
sns.lineplot(x='avg_emp_rating',y='num_emp',data=df)
plt.show()
```

```python
sns.lineplot(x='avg_emp_rating',y='avg_emp_tenure_in_years', data=df)
sns.lineplot(x='avg_emp_rating',y='num_emp',data=df)
plt.legend(labels=['Average # of years','# of employees'])
plt.show()
```

```python
sns.lineplot(x='avg_emp_rating',y='avg_emp_tenure_in_years', data=df)
sns.lineplot(x='avg_emp_rating',y='num_emp',data=df)
plt.legend(labels=['Average number of year','Number of employees'])
plt.xlabel('Average employee rating')
plt.ylabel('Number')
plt.show()
```
