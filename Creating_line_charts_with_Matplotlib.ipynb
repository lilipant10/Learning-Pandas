```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('https://raw.githubusercontent.com/Thinkful-Ed/data-science-lectures/master/salary_data.csv')
df
```

```python
plt.plot(df['Year'],df['Assoc'])
plt.show()
```

```python
plt.plot(df['Year'], df['Assoc'])
plt.plot(df['Year'], df['Spv'])
plt.plot(df['Year'], df['Mgr'])
plt.show()
```

```python
plt.plot(df['Year'], df['Assoc'])
plt.plot(df['Year'], df['Spv'])
plt.plot(df['Year'], df['Mgr'])
plt.legend(labels=['Associate','Supervisor','Manager'])
plt.show()
```
